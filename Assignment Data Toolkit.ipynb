```python
#14.Create time series dataset in pandas dataframe with column Date,Temperature,Humidity and plot
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dates = pd.date_range(start='2024-01-01', periods=100, freq='D')

temperature = np.random.uniform(low=-10, high=35, size=len(dates))  # Temperature between -10 and 35 degrees
humidity = np.random.uniform(low=20, high=90, size=len(dates))     # Humidity between 20% and 90%



df = pd.DataFrame({
    'Date': dates,
    'Temperature': temperature,
    'Humidity': humidity
})

print(df.head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(30,10))                   #makes a figure

plt.suptitle('Temperature and Humidity over Time',size=50)
plt.subplot(1,2,1)
plt.plot(dates,temperature)
plt.xlabel('Date',size=20)

plt.subplot(1,2,2)
plt.plot(dates,humidity)
plt.xlabel('Date',size=20)


plt.show()
```

```python
 #15.Create numpy array containing1000 samples from normal distribution and plot
```

```python
from scipy.stats import norm
samples=np.random.randn(1000)
x=norm.pdf(samples,loc=np.mean(samples),scale=np.std(samples))
plt.hist(samples,bins=30)
plt.plot(samples,x,linewidth=10)
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title("Histogram with PDF overlay")
plt.show()
```

```python
#17.Create a seaborn scatter plot of two random arrays, color points based on position relative to quadrant,add legend
#   label axes,and title=Quadrant wiseScatter Plot
```

```python
import seaborn as sns
x=np.random.randn(100)
y=np.random.randn(100)

quadrants = []
for i in range(len(x)):                         #100 is the size of x,y
    if x[i] > 0 and y[i] > 0:
        quadrants.append('Q1')
    elif x[i] < 0 and y[i] > 0:
        quadrants.append('Q2')
    elif x[i] < 0 and y[i] < 0:
        quadrants.append('Q3')
    else:
        quadrants.append('Q4')



data=pd.DataFrame({'x':x,'y':y,'Quad':quadrants})
plt.figure()
sns.scatterplot(data=data,x='x', y='y', hue='Quad')

plt.title('Quadrant-wise Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')


plt.legend(title='Quadrant')
```

```python
#18.With Bokeh, plot a line chart of a sine wave function, add grid lines, label axes, set the title Sine Wave Function
```

```python
#19.Using Bokeh generate a bar chart of randomly generated categorical data, color bars based on their
#   values, add hover tooltips to display values, label axes,set the title as Random Categorical Bar Chart
```

```python
import bokeh.io
import bokeh.plotting

from bokeh.plotting import figure,output_file,show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Viridis256


categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [np.random.randint(1, 100) for _ in categories]
# Create a ColumnDataSource
colors = ['blue', 'green', 'red', 'purple', 'orange']
source = ColumnDataSource(data=dict(categories=categories, values=values, color=colors))

p = figure(x_range=categories, title="Random Categorical Bar Chart", tools="")

# Add bars with colors
p.vbar(x='categories', top='values', source=source,color='color', legend_field='categories')

# Add hover tool
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)


p.xaxis.axis_label = "Categories"
p.yaxis.axis_label = "Values"



show(p)
```
